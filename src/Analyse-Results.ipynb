```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread

%matplotlib ipympl
```

```python
f_name = '0801'
cmap = 'gist_heat'
vmax = 2

f_hr = '../../Data/DIV2K_TIFF/DIV2K_valid_HR/{}.tiff'
f_sr = '../experiment/edsrX2_DIV2K_TIFF/results-DIV2K_TIFF/{}_x2_SR.tiff'

img_hr = imread(f_hr.format(f_name))
img_sr = imread(f_sr.format(f_name))

diff = np.abs(img_hr - img_sr)
diff[diff >= 0.1] = 1

plt.figure()
plt.imshow(diff, cmap=cmap)
plt.grid(False)
plt.axis('off')
plt.colorbar()
plt.show()
```
